```python
import os
import cv2
import json

# Define the paths to the image directory and annotation file
image_dir = '/content/archive (10)/Apply_Grayscale/Apply_Grayscale/Vehicles_Detection.v9i.coco/test'  # Update to your image directory
annotation_file = '/content/archive (10)/Apply_Grayscale/Apply_Grayscale/Vehicles_Detection.v9i.coco/test/_annotations.coco.json'  # Path to annotation JSON

# Function to verify if an image exists and can be read
def verify_images(image_dir):
    missing_images = []
    for filename in os.listdir(image_dir):
        if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
            img_path = os.path.join(image_dir, filename)
            img = cv2.imread(img_path)
            if img is None:
                missing_images.append(filename)

    return missing_images

# Function to verify if the annotations file exists and is valid
def verify_annotations(annotation_file):
    if not os.path.exists(annotation_file):
        print(f"Annotation file {annotation_file} does not exist.")
        return False

    try:
        with open(annotation_file, 'r') as f:
            annotations = json.load(f)  # Try to load the JSON file
        return annotations
    except json.JSONDecodeError:
        print(f"Failed to decode JSON from {annotation_file}")
        return False

# Check image integrity
missing_images = verify_images(image_dir)
if missing_images:
    print("Missing or unreadable images:")
    for img in missing_images:
        print(img)
else:
    print("All images are valid!")

# Check annotation integrity
annotations = verify_annotations(annotation_file)
if annotations:
    print("Annotation file loaded successfully.")
else:
    print("Annotation file verification failed.")
```

```python
import os

def verify_image_annotation_pairs(image_dir, label_dir):
    image_files = os.listdir(image_dir)
    label_files = os.listdir(label_dir)

    # Ensure all image files have corresponding label files
    for image_file in image_files:
        # Extract the base name without extension (e.g., 'frame_0002_jpg.rf.a929cd43bfa0953ace4ea9ea00fa9ba7')
        base_name = os.path.splitext(image_file)[0]

        # Check if the corresponding label file exists
        label_file = base_name + '.txt'

        if label_file not in label_files:
            print(f"[ERROR] No label file found for image: {image_file}")
        else:
            print(f"Image: {image_file} has a corresponding label: {label_file}")

    # Check if all label files have a corresponding image
    for label_file in label_files:
        base_name = os.path.splitext(label_file)[0]
        image_file = base_name + '.jpg'

        if image_file not in image_files:
            print(f"[ERROR] No image found for label: {label_file}")
        else:
            print(f"Label: {label_file} has a corresponding image: {image_file}")

# Example usage
image_dir = '/content/archive (10)/No_Apply_Grayscale/No_Apply_Grayscale/Vehicles_Detection.v8i.yolov8/train/images'
label_dir = '/content/archive (10)/No_Apply_Grayscale/No_Apply_Grayscale/Vehicles_Detection.v8i.yolov8/train/labels'

verify_image_annotation_pairs(image_dir, label_dir)
```

```python
import cv2
import os
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

def load_and_preprocess_images(image_dir, target_size=(416, 416)):
    images = []
    labels = []
    for label in os.listdir(image_dir):
        label_path = os.path.join(image_dir, label)
        if os.path.isdir(label_path):
            for img_file in os.listdir(label_path):
                if img_file.endswith('.jpg'):
                    img_path = os.path.join(label_path, img_file)
                    # Read and resize image
                    img = cv2.imread(img_path)
                    img = cv2.resize(img, target_size)  # Resize image to 416x416
                    img = img / 255.0  # Normalize image to [0, 1]
                    images.append(img)
                    labels.append(label)  # You might need to convert label to integer
    return np.array(images), np.array(labels)

# Example: Load train images
train_image_dir = '/content/archive (10)/Apply_Grayscale/Apply_Grayscale/Vehicles_Detection.v9i.yolov8/train/images'
X_train, y_train = load_and_preprocess_images(train_image_dir)

# Example: Load validation images
valid_image_dir = '/content/archive (10)/Apply_Grayscale/Apply_Grayscale/Vehicles_Detection.v9i.yolov8/valid/images'
X_valid, y_valid = load_and_preprocess_images(valid_image_dir)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

def create_cnn_model(input_shape=(416, 416, 3), num_classes=3):
    model = Sequential()

    # First convolutional block
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))

    # Second convolutional block
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    # Third convolutional block
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    # Flatten and add fully connected layers
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))  # Dropout to prevent overfitting
    model.add(Dense(num_classes, activation='softmax'))  # Softmax for multi-class classification

    # Compile the model
    model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model

# Create the CNN model
num_classes = len(np.unique(y_train))  # Number of unique classes in your dataset
model = create_cnn_model(num_classes=num_classes)

# Summary of the model
model.summary()
```

```python
import os
import cv2
import numpy as np

def load_images_and_labels(image_dir, label_dir, image_size=(416, 416)):
    images = []
    labels = []
    for img_file in os.listdir(image_dir):
        if img_file.endswith('.jpg'):
            img_path = os.path.join(image_dir, img_file)
            label_file = img_file.replace('.jpg', '.txt')
            label_path = os.path.join(label_dir, label_file)

            # Check if the label file exists
            if not os.path.exists(label_path):
                print(f"Warning: Label file for {img_file} is missing!")
                continue  # Skip this image if the label is missing

            # Load and process the image
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error: Unable to read image {img_file}")
                continue  # Skip if the image cannot be read
            img = cv2.resize(img, image_size) / 255.0  # Normalize image
            images.append(img)

            # Read the label file
            with open(label_path, 'r') as f:
                label = []
                for line in f.readlines():
                    class_id, x, y, w, h = map(float, line.strip().split())
                    label.append([class_id, x, y, w, h])
                labels.append(np.array(label))

    # Convert lists to numpy arrays
    return np.array(images), labels

# Example usage
train_image_dir = '/content/archive (10)/Apply_Grayscale/Apply_Grayscale/Vehicles_Detection.v9i.yolov8/train/images'
train_label_dir = '/content/archive (10)/Apply_Grayscale/Apply_Grayscale/Vehicles_Detection.v9i.yolov8/train/labels'

X_train, y_train = load_images_and_labels(train_image_dir, train_label_dir)

# Check the shape of the output arrays
print(f"Loaded {len(X_train)} images and {len(y_train)} labels.")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, GlobalAveragePooling2D, Input
from tensorflow.keras.optimizers import Adam

# Define the model with an Input layer
model = Sequential()

# Input layer
model.add(Input(shape=(224, 224, 3)))  # Reduced image size (224x224)

# Convolutional layers for feature extraction
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Use Global Average Pooling instead of Flatten
model.add(GlobalAveragePooling2D())

# Dense layers with reduced units
model.add(Dense(64, activation='relu'))

# Output layer: 5 bounding boxes, each with 5 values (4 coordinates + 1 class)
max_boxes = 5
model.add(Dense(max_boxes * 5, activation='sigmoid'))  # 5 boxes, each with 5 values

# Compile the model
model.compile(optimizer=Adam(), loss='mean_squared_error', metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Assume you have your ground truth and predictions
y_true = [0, 1, 1, 0]  # True labels
y_scores = [0.1, 0.4, 0.35, 0.8]  # Predicted scores (probabilities or confidence)

precision, recall, _ = precision_recall_curve(y_true, y_scores)

plt.plot(recall, precision, label="Precision-Recall curve")
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision vs Recall')
plt.legend()
plt.show()
```
